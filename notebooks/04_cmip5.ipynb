```python
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
%matplotlib inline
import cartopy.crs as ccrs
```

# Daily output

```python
plt.figure(figsize=(15,8))

ax = plt.subplot(3,1,1,projection=ccrs.PlateCarree())
ds0 = xr.open_dataset('../data/raw/tas_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
ds = xr.open_dataset('../data/raw/tas_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
p=ds['tas'].sel(time=ds['time'][195]).plot(transform=ccrs.PlateCarree())
ax.coastlines()
p.set_clim([280,320])
c=ds['tas'].sel(time=ds['time'][195]).plot.contour(levels=[310],linewidths=2,colors='k',transform=ccrs.PlateCarree())
plt.ylim([-30,30])

ax = plt.subplot(3,1,2,projection=ccrs.PlateCarree())
ds = xr.open_dataset('../data/raw/clt_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
ds['clt'].sel(time=ds['time'][195]).plot(transform=ccrs.PlateCarree())
ax.coastlines()
c=ds0['tas'].sel(time=ds0['time'][195]).plot.contour(levels=[310],linewidths=2,colors='k',transform=ccrs.PlateCarree())
plt.ylim([-30,30])

ax = plt.subplot(3,1,3,projection=ccrs.PlateCarree())
ds = xr.open_dataset('../data/raw/hur_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
ds['hur'].sel(time=ds['time'][195],plev=ds['plev'][3]).plot(transform=ccrs.PlateCarree())
ax.coastlines()
c=ds0['tas'].sel(time=ds0['time'][195]).plot.contour(levels=[310],linewidths=2,colors='k',transform=ccrs.PlateCarree())
plt.ylim([-30,30])

plt.savefig('../figures/4xCO2_07-15_fields.png',bbox_inches='tight',dpi=200)
```

```python
plt.figure(figsize=(15,8))

ax = plt.subplot(3,1,1,projection=ccrs.PlateCarree())
ds0 = xr.open_dataset('../data/raw/tas_day_GFDL-ESM2M_piControl_r1i1p1_02960101-03001231.nc')
ds = xr.open_dataset('../data/raw/tas_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
p=(ds['tas']-ds0['tas']).mean(dim='time').plot(transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu_r'))
ax.coastlines()
p.set_clim([-10,10])
plt.ylim([-30,30])

ax = plt.subplot(3,1,2,projection=ccrs.PlateCarree())
ds0 = xr.open_dataset('../data/raw/clt_day_GFDL-ESM2M_piControl_r1i1p1_02960101-03001231.nc')
ds = xr.open_dataset('../data/raw/clt_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
(ds['clt']-ds0['clt']).mean(dim='time').plot(transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu_r'))
ax.coastlines()
plt.ylim([-30,30])

ax = plt.subplot(3,1,3,projection=ccrs.PlateCarree())
ds0 = xr.open_dataset('../data/raw/hur_day_GFDL-ESM2M_piControl_r1i1p1_02960101-03001231.nc')
ds = xr.open_dataset('../data/raw/hur_day_GFDL-ESM2M_abrupt4xCO2_r1i1p1_02960101-03001231.nc')
(ds['hur']-ds0['hur']).sel(plev=ds['plev'][3]).mean(dim='time').plot(transform=ccrs.PlateCarree(),cmap=plt.get_cmap('RdBu_r'))
ax.coastlines()
plt.ylim([-30,30])

plt.savefig('../figures/4xCO2_5yr-mean_change.png',bbox_inches='tight',dpi=200)
```
